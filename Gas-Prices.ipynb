```python
#Some notable observations: The mean nominal disposable income was $3.171x10^6; the mean gasoline retial price was about 96 cents'gallon.
```

```python
from sklearn.linear_model import LinearRegression
slr = LinearRegression()
X = gas[['X3']].values
y = gas[['Y']].values
slr.fit(X, y)

print ('Slope: %3f' % slr.coef_[0])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def lin_regplot(X, y, model):
    plt.scatter(X, y, c = 'steelblue', edgecolor = 'white', s=70)
    plt.plot(X, model.predict(X), color = 'black', lw = 2)
    return None

lin_regplot(X, y, slr)
plt.xlabel('EDUC')
plt.ylabel('SALAY')
plt.show()
```

```python
gas['lg(SGAS)'] = np.log(gas['SGAS'])
gas
```

```python
gas['lg(PGAS)'] = np.log(gas['PGAS'])
gas
```

```python
from sklearn.linear_model import LinearRegression
slr = LinearRegression()
X = gas[['lg(PGAS)']].values
y = gas[['lg(SGAS)']].values
slr.fit(X, y)

print ('Slope: %3f' % slr.coef_[0])
```

```python
def lin_regplot(X, y, model):
    plt.scatter(X, y, c = 'purple', edgecolor = 'white', s=70)
    plt.plot(X, model.predict(X), color = 'black', lw = 2)
    return None

lin_regplot(X, y, slr)
plt.xlabel('lg(PGAS)')
plt.ylabel('lg(SGAS)')
plt.show()
```
